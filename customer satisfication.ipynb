```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("https://github.com/AshwiniDeva2006/keydrivers/raw/main/keydrivers.csv")  # Adjust path if needed

# Basic info
print(df.head())
print(df.info())
print(df.describe())
print("Missing Values:\n", df.isnull().sum())
print("Duplicate Records:", df.duplicated().sum())
# 1. Satisfaction Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Overall Satisfaction', data=df, palette='viridis')
plt.title("Overall Satisfaction Distribution")
plt.xlabel("Satisfaction Score")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# 2. Platform-wise Average Satisfaction
plt.figure(figsize=(7, 4))
sns.barplot(x='Platform', y='Overall Satisfaction', data=df, ci=None, palette='Set2')
plt.title("Average Satisfaction by Platform")
plt.ylabel("Avg. Satisfaction")
plt.tight_layout()
plt.show()
# 4. Customer Support by Satisfaction Level
plt.figure(figsize=(7, 5))
sns.boxplot(x='Overall Satisfaction', y='Customer Support', data=df, palette='pastel')
plt.title("Customer Support vs Satisfaction Level")
plt.tight_layout()
plt.show()

# 5. Summary insights from groupby operations
print("\n📊 Average Satisfaction by Platform:\n", df.groupby('Platform')['Overall Satisfaction'].mean())
print("\n📊 Correlation with Overall Satisfaction:\n", df.drop(columns=['ID', 'Platform']).corr(numeric_only=True)['Overall Satisfaction'].sort_values(ascending=False))

# 6. Key metric highlights
top_feature = df.drop(columns=['ID', 'Platform']).corr(numeric_only=True)['Overall Satisfaction'].drop('Overall Satisfaction').idxmax()
print(f"\n⭐ Top Positive Correlating Feature: {top_feature}")

# 7. Summary of Customer Support by Satisfaction Level
print("\n📊 Customer Support Summary by Satisfaction Level:\n", df.groupby('Overall Satisfaction')['Customer Support'].describe())
```
